## Implementing Crow Agent's Responsibilities

## Generator Phase

```python
from mini_robin.configuration import RobinConfiguration
from mini_robin.file_system import FileSystem

from mini_robin.hypotheses_generator import HypothesesGenerator
from mini_robin.hypotheses_ranker import HypothesesRanker

from mini_robin.mini_robin_crew import MiniRobinAssayCrew, MiniRobinRankingCrew

from mini_robin.bedrock_llm_wrapper import BedrockLLMProvider

robin_config = RobinConfiguration(
    disease_name = "glaucoma",
    num_queries = 5,
    num_assays = 5,
    num_candidates = 2,
)


# --- Hypotheses generation Phase ----

hypotheses_generator_llm = BedrockLLMProvider(model_name="bedrock/meta.llama3-70b-instruct-v1:0").get_llm()

mock_knowledge_source_glucoma =["/Users/sarangsonar/Documents/GitHub/mini_robin/pdf/glaucoma/glaucoma_1.pdf","/Users/sarangsonar/Documents/GitHub/mini_robin/pdf/glaucoma/glaucoma_2.pdf",
       "/Users/sarangsonar/Documents/GitHub/mini_robin/pdf/glaucoma/glaucoma_3.pdf"]

assay_crew_runner = MiniRobinAssayCrew(llm=hypotheses_generator_llm, files = mock_knowledge_source_glucoma, file_system = FileSystem(robin_config))



hypotheses_generator = HypothesesGenerator(robin_config = robin_config,
                                           generator_assay_crew_instance = assay_crew_runner)

hypotheses_results = hypotheses_generator.generate()
hypotheses_generator.save_df()




```

```python
hypotheses_generator.to_dataframe()
```

## Ranking Phase

```python

# --- Hypotheses ranking Phase ----

hypotheses_ranker_llm_judge = BedrockLLMProvider(model_name="bedrock/meta.llama3-70b-instruct-v1:0").get_llm()

ranking_crew_runner = MiniRobinRankingCrew(llm=hypotheses_ranker_llm_judge)

# hypothese_dataframe = load_hypotheses_json_to_df("/Users/sarangsonar/Documents/GitHub/mini_robin/mini_robin/robin_output/glucoma/hypotheses_outputs/hypotheses.json")

ranker = HypothesesRanker(
    robin_config =  robin_config,
    ranker_crew_instance = ranking_crew_runner,
    # hypotheses_dataframe = hypothese_dataframe,
    file_system = FileSystem(robin_config),
    ranking_system = "elo"
    
)


ranking_results = ranker.run_all_comparison()
ranker.save_df()
```
